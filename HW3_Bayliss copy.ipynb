Create grids

```python
def create_grids(n, alpha, beta, delta, A):

    zero_grid=np.linspace(0, 1, n).reshape(-1,1)

    k_grid_1=np.zeros(shape=(n,1))

    k_grid_2=np.zeros(shape=(n,1))

    k_grid_3=np.zeros(shape=(n,1))

    k_grid_4=np.zeros(shape=(n,1))

    k_ss  = (( alpha * beta * A) / (1 - beta * (1 - delta)))** (1/(1 - alpha))
    
    #for i in range(len(new_grid)):
        #k_grid[i] =  low_bound + ( up_bound - low_bound ) * (new_grid[i] ** 1)

    for i in range(len(zero_grid)):
        k_grid_1[i] =  (0.75 * k_ss) + ( (1.25 * k_ss) - (0.75 * k_ss) ) * (zero_grid[i] ** 1)

    for i in range(len(zero_grid)):
        k_grid_2[i] =  (0.75 * k_ss) + ( (1.25 * k_ss) - (0.75 * k_ss) ) * (zero_grid[i] ** 1.5)

    for i in range(len(zero_grid)):
        k_grid_3[i] =  (0.75 * k_ss) + ( (1.25 * k_ss) - (0.75 * k_ss) ) * (zero_grid[i] ** 3)

    for i in range(len(zero_grid)):
        k_grid_4[i] =  (0.75 * k_ss) + ( (1.25 * k_ss) - (0.75 * k_ss) ) * (zero_grid[i] ** 4)

    return k_grid_1, k_grid_2, k_grid_3, k_grid_4
```

Utility

```python
def u(c, gamma):
    if gamma == 1:
        return np.log(c)
    else:
        return ((c ** (1 - gamma))/( 1 - gamma))
```

```python
def u_new(c,gamma, beta):
    return (1 - beta) * (c) ** (1-gamma)
```

Production

```python
def f(A, alpha, delta, k):
    return A * (k ** alpha) + (1 - delta) * k
```

# First, let us use VFI with a discrete grid as a review

```python
def Tv(v, k, delta, toler, max_iter, alpha, A, beta, n, gamma):
    start_time = time.time()
    iter = 1
    error = toler+1
    v1=np.zeros(shape=(n,1))
    policy=np.zeros(shape=(n,1))
    while (toler<error) and (iter<max_iter):
        distance=np.zeros(shape=(n,1)) 
        for i in range(len(k)):
            vmax = -np.inf
            for j in range(len(k)): 
                c = f(A, alpha, delta, k[i]) - k[j] 
                if c>=0: 
                    v_temp = u(c, gamma)+beta*v[j]
                    if v_temp>vmax: 
                        vmax=v_temp 
                        policy_i=k[j]
                v1[i]=vmax 
                policy[i]=policy_i 
                distance[i]=np.abs(vmax-v[i]) 
        error=np.max(distance)       
        v=np.copy(v1) 
        if iter == 1:
            print(f"Iteration {iter}, Max Error: {error}") 
        if iter % 10 == 0:
            print(f"Iteration {iter}, Max Error: {error}") 
        iter+=1
    if iter == max_iter:
        print("Failed to converge in max_iter iterations") 
    end_time = time.time() 
    execution_time = end_time - start_time 
    print(f"Execution time: {execution_time} seconds")
    return v, policy 
```

# Now, we interpolate

Deal with the case of log utility first

```python
def T(toler, max_iter, beta, gamma, alpha, delta, A, n):

    V = np.zeros(shape=(n,1))

    V_new = np.zeros(shape=(n,1))

    G_kp = np.zeros(shape=(n,1))

    k = create_grids(n, alpha, beta, delta, A)[3]

    iter = 1

    start_time = time.time()

    for iter in range(max_iter):

        for i, k_i in enumerate(k):

            def F(kp, k_i, V_interp):
            # Objective Function
            # Calculate consumption for each kp
                c = f(A, alpha, delta, k_i) - kp
                # Minimise the negative of the Bellman with interpolated v(kp)
                return -u(c, gamma) - beta * V_interp(kp)
        
        # Interpolate the value function

            V_interp = interpolate.interp1d(np.squeeze(k), np.squeeze(V) , kind = 'cubic', fill_value = 'extrapolate')

            k_min = np.min(k)

            k_max = f(A, alpha, delta, k[i])

            # Derivative of the objective function at the lower bound

            dF_min = -1

            # Derivative of the objective function at the upper bound

            dF_max = 1

            # Check for corner solutions
            # If the derivative at kp at the lower bound is negative, it suggests that the objective function is decreasing as kp approaches the lower bound
            # This can happen if the policy implied by the objective function is to reduce kp further.
            # So, the algorithm sets kp_min to k_min, this implies that if the optimisation algorithm is in a situation
            # where it seems beneficial to reduce kp below k_min based on the objective function, the algorithm starts the search from k_min.
            # The idea is to handle scenarios where the objective function might suggest a corner solution at the lower bound. 
            # Setting kp_min to k_min provides a starting point for the optimisation algorithm.

            # While kp_min and kp_max are indeed set to k_min and k_max respectively, they are used as initial guesses for the minimisation algorithm. 
            # However, these initial guesses do not restrict the search space of the algorithm; 
            # the algorithm is still free to explore values within the bounds specified in the bounds parameter of minimize_scalar.

            kp_min = k_min if dF_min < 0 else None

            kp_max = k_max if dF_max > 0 else None

            # Since we pass args=(k_i, V_interp), minimize_scalar will search in the kp direction to minimise F.
            # That is, it will find consumption for each kp, and calculate the Bellman using V_interp(kp)
            # Result will be of type OptimizeResult
            # To find the value, we use -result.fun, and result.x is the optimal value of kp which minimises the objective function.
            # Minimise_scalar uses two-sided derivatives

            result = minimize_scalar(F, bounds=(kp_min, kp_max), args=(k_i, V_interp), method='bounded')

            G_kp[i] = result.x

            V_new[i] = -result.fun
        
        iter += 1

        error = np.abs(np.max((V_new - V)/(1+np.abs(V))))

        if iter == 1:
            print(f"Iteration {iter}, Max Error: {error}") 
        if iter % 25 == 0:
            print(f"Iteration {iter}, Max Error: {error}") 

        if np.max(np.abs(V_new - V)/(1+np.abs(V))) < toler:
            print(f"Converged in {iter} iterations")
            break

        V = np.copy(V_new)
    
    end_time = time.time() 
    execution_time = end_time - start_time 
    print(f"Execution time: {execution_time} seconds")

    # Plots

    a0 = (np.log(A) + (1 - alpha * beta) * np.log(1 - alpha * beta) + alpha * beta * np.log(alpha * beta)) / ((1 - beta) * (1 - alpha * beta))

    a1 = alpha / (1 - alpha * beta)

    plt.plot(k, a0 + a1 * np.log(k), label='exact')
    plt.plot(k, V_interp(k), label='vfi')
    plt.xlabel('k')
    plt.ylabel("v")
    plt.title("Value function")
    plt.legend()
    plt.show()

    plt.plot(k, A * alpha * beta * k ** alpha, label='exact')
    plt.plot(k, G_kp, label='vfi')
    plt.xlabel('k')
    plt.ylabel("v")
    plt.title("Policy function")
    plt.legend()
    plt.show()


    return V, G_kp            
```

```python
out_valu, out_poli = T(toler=1e-7, max_iter=10000, beta=0.95, gamma=1, alpha=0.33, delta=1, A=10, n=31)
```

Now, fix the code so that we can use CRRA, `gamma=2`. Note that in the following code, we use Epstein-Zin preferences. These are given by,

$ V(k) = \underset{c,k'}{\textrm{max}} \{(1-\beta)c^{1-\gamma}+\beta V(k')^{1-\gamma}\}^{\frac{1}{1-\gamma}} $

```python
def Tw(toler, max_iter, beta, gamma, alpha, delta, A, n):

    V = np.ones(shape=(n,1))

    V_new = np.zeros(shape=(n,1))

    G_kp = np.zeros(shape=(n,1))

    k = create_grids(n, alpha, beta, delta, A)[3]

    iter = 1

    start_time = time.time()

    for iter in range(max_iter):

        for i, k_i in enumerate(k):

            def F(kp, k_i, V_interp):
                c = f(A, alpha, delta, k_i) - kp
                return -((u_new(c, gamma, beta) + beta * (V_interp(kp))**(1-gamma)))**(1/(1-gamma))

            V_interp = interpolate.interp1d(np.squeeze(k), np.squeeze(V) , kind = 'cubic', fill_value = 'extrapolate')

            k_min = np.min(k)

            k_max = f(A, alpha, delta, k[i])

            dF_min = -1

            dF_max = 1

            kp_min = k_min if dF_min < 0 else None

            kp_max = k_max if dF_max > 0 else None

            result = minimize_scalar(F, bounds=(kp_min, kp_max), args=(k_i, V_interp), method='bounded')

            G_kp[i] = result.x

            V_new[i] = -result.fun
        
        iter += 1

        error = np.abs(np.max((V_new - V)/(1+np.abs(V))))

        if iter == 1:
            print(f"Iteration {iter}, Max Error: {error}") 
        if iter % 25 == 0:
            print(f"Iteration {iter}, Max Error: {error}") 

        if np.max(np.abs(V_new - V)/(1+np.abs(V))) < toler:
            print(f"Converged in {iter} iterations")
            break

        V = np.copy(V_new)

    end_time = time.time() 
    execution_time = end_time - start_time 
    print(f"Execution time: {execution_time} seconds")

    # Plots

    plt.plot(k, V_interp(k), label='vfi')
    plt.xlabel('k')
    plt.ylabel("v")
    plt.title("Value function")
    plt.legend()
    plt.show()

    plt.plot(k, G_kp, label='vfi')
    plt.xlabel('k')
    plt.ylabel("v")
    plt.title("Policy function")
    plt.legend()
    plt.show()


    return V, G_kp            
```

Finally, we run the code using the following preferences,

$ V(k) = \underset{c,k'}{\textrm{max}} \{\frac{c^{1-\gamma}}{1-\gamma}+ \beta V(k')\} $


```python
def Tq(toler, max_iter, beta, gamma, alpha, delta, A, n):

    V = np.ones(shape=(n,1))

    V_new = np.zeros(shape=(n,1))

    G_kp = np.zeros(shape=(n,1))

    k = create_grids(n, alpha, beta, delta, A)[3]

    iter = 1

    start_time = time.time()

    for iter in range(max_iter):

        for i, k_i in enumerate(k):

            def F(kp, k_i, V_interp):
                c = f(A, alpha, delta, k_i) - kp
                return - ((u(c, gamma) + beta * (V_interp(kp))))

            V_interp = interpolate.interp1d(np.squeeze(k), np.squeeze(V) , kind = 'cubic', fill_value = 'extrapolate')

            k_min = np.min(k)

            k_max = f(A, alpha, delta, k[i])

            dF_min = -1

            dF_max = 1

            kp_min = k_min if dF_min < 0 else None

            kp_max = k_max if dF_max > 0 else None

            result = minimize_scalar(F, bounds=(kp_min, kp_max), args=(k_i, V_interp), method='bounded')

            G_kp[i] = result.x

            V_new[i] = -result.fun
        
        iter += 1

        error = np.abs(np.max((V_new - V)/(1+np.abs(V))))

        if iter == 1:
            print(f"Iteration {iter}, Max Error: {error}") 
        if iter % 25 == 0:
            print(f"Iteration {iter}, Max Error: {error}") 

        if np.max(np.abs(V_new - V)/(1+np.abs(V))) < toler:
            print(f"Converged in {iter} iterations")
            break

        V = np.copy(V_new)

    end_time = time.time() 
    execution_time = end_time - start_time 
    print(f"Execution time: {execution_time} seconds")

    # Plots

    plt.plot(k, V_interp(k), label='vfi')
    plt.xlabel('k')
    plt.ylabel("v")
    plt.title("Value function")
    plt.legend()
    plt.show()

    plt.plot(k, G_kp, label='vfi')
    plt.xlabel('k')
    plt.ylabel("v")
    plt.title("Policy function")
    plt.legend()
    plt.show()


    return V, G_kp  
```

# Comment on convergence criteria

Above, we defined our convergence criteriaon the value function. What happens if we define this on the policy function instead? After all, the policy function is what we are really interested in. It is important to note that we are only checking for convergence ON grid points. This can lead to issues, but for now, we naively run with it.

It is also worth bearing in mind that the value function will almost certainly be "incorrect" when defining convergence in this way.

# Now, let us start to add uncertainty to the model

First, let us discretise the AR1 process for TFP. To do so, we will use the Rouwenhorst method from the QuantEcon package. The output will consist of the shock values, and the associated transition matrix.

One thing to note is that in this first attempt at adding shocks, we will get a value function for each possible shock, rather than a 3D value function.

Also, we need to create new grids. Let's do that first.

```python
def create_grids_stochastic(n):

    zero_grid=np.linspace(0, 1, n).reshape(-1,1)

    k_grid_1=np.zeros(shape=(n,1))

    k_grid_2=np.zeros(shape=(n,1))

    k_grid_3=np.zeros(shape=(n,1))

    k_grid_4=np.zeros(shape=(n,1))

    k_ss_low  = 0.01

    k_ss_high  = 2
    
    #for i in range(len(new_grid)):
        #k_grid[i] =  low_bound + ( up_bound - low_bound ) * (new_grid[i] ** 1)

    for i in range(len(zero_grid)):
        k_grid_1[i] =  (0.25 * k_ss_low) + ( (1.5 * k_ss_high) - (0.75 * k_ss_low) ) * (zero_grid[i] ** 1)

    for i in range(len(zero_grid)):
        k_grid_2[i] =  (0.25 * k_ss_low) + ( (1.5 * k_ss_high) - (0.75 * k_ss_low) ) * (zero_grid[i] ** 1.5)

    for i in range(len(zero_grid)):
        k_grid_3[i] =  (0.25 * k_ss_low) + ( (1.5 * k_ss_high) - (0.75 * k_ss_low) ) * (zero_grid[i] ** 3)

    for i in range(len(zero_grid)):
        k_grid_4[i] =  (0.25 * k_ss_low) + ( (1.5 * k_ss_high) - (0.75 * k_ss_low) ) * (zero_grid[i] ** 4)

    return k_grid_1, k_grid_2, k_grid_3, k_grid_4
```

Let's first solve the log utility case.

```python
def model_corrected(toler, max_iter, beta, gamma, alpha, delta, A, rho, sigma, n_k, n_z, mu):

    # Discretize TFP shocks using Rouwenhorst
    mc = qe.markov.approximation.rouwenhorst(n_z, rho, sigma, mu)
    z, Pi = mc.state_values, mc.P 
    
    # Create grids for capital and TFP shock
    z_grid = z
    k_grid = create_grids_stochastic(n_k)[1]

    # Initialize value function as a 2D array
    V = np.zeros((n_k, n_z))

    G_kp = np.zeros((n_k, n_z))

    G_kp_old = np.zeros((n_k, n_z))

    it = 0

    for it in range(max_iter):

        V_new = np.zeros_like(V)
        G_kp = np.zeros_like(V)

        for i, z_i in enumerate(z_grid):

            Ev = np.dot(V, Pi[i,:])
            
            IEv = interpolate.interp1d(np.squeeze(k_grid), np.squeeze(Ev), kind='cubic', fill_value='extrapolate')
            
            for j, k_j in enumerate(k_grid):

                # Bellman equation
                def B(kp, k_j, z_i, IEv):
                    c = (k_j ** alpha) * np.exp(A * z_i) + (1 - delta) * k_j - kp
                    return -((u(c, gamma)) + beta * IEv(kp))
                
                k_min = 0

                k_max = f(np.exp(A * z_i), alpha, delta, k_j)   
                 
                # k_max = 20            

                # Find optimal capital stock for the given TFP shock
                result = minimize_scalar(B, args=(k_j, z_i, IEv), bounds=(k_min, k_max), method='bounded')

                # Update policy function
                G_kp[j,i] = result.x

                # Update value function
                V_new[j, i] = -result.fun

        # Check for convergence using the scaled error
        error = np.max(np.abs(G_kp - G_kp_old) / (1 + np.abs(G_kp_old)))

        if it % 10 == 0:
            print(f"Iteration {it}, Max Error: {error}")

        if error < toler:
            print(f"Converged in {it + 1} iterations")
            break

        V = np.copy(V_new)

        G_kp_old = np.copy(G_kp)

        it += 1

    fig, ax = plt.subplots(figsize=(10, 6))
    for j in range(n_z):
        ax.plot(k_grid, V[:, j], label=f'TFP Shock {j}')

    ax.set_title('Value Function')
    ax.set_xlabel('Capital Stock')
    ax.set_ylabel('Value')
    ax.legend()
    plt.show()

    # Plot the policy function
    fig, ax = plt.subplots(figsize=(10, 6))
    for j in range(n_z):
        ax.plot(k_grid, G_kp[:, j], label=f'TFP Shock {j}')

    ax.set_title('Policy Function')
    ax.set_xlabel('Current Capital Stock')
    ax.set_ylabel('Optimal Next Period Capital Stock')
    ax.legend()
    plt.show()

    return G_kp, V
```
